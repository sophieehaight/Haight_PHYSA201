```python
#import libraries/packages
from astropy import units as u
from astropy import constants as const 
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining constants here used throughout the assignment:

#in the future, I'll try to use astropy for these
k_B = 1.3807*10**(-16)
c = 3.0*10**(10)
h= 6.626*10**(-27)

#define distance from source and wavelength ranges which will be the same throughout in this assignment
d= 3.0*10**19 #10pc in cm
lambda_min = 10**(-5) #in cm
lambda_max= 10**(-2) #in cm
```

```python
#black body calculation function
def bb_lambda(T, R, d, l_min, l_max,title, logplot=True):
    #define a range of wavelengths to plot between l_min and l_max
    
    #set step size assuming were plotting in microns
    wavelengths = np.arange(l_min, l_max, 0.01*10**(-4))
    
    
    #init an empty list for fluxes:
    fluxes = []
    #for loop to calculate a flux for each wavelength:
    for i in wavelengths:
        #define a scaling factor for distance from source:
        dist_scale = (R/d)**2
        
        #define a flux per v and T, for an object with radius R, at a distance d
        fluxes.append(float(dist_scale*(2*h*c**2/i**5)/(np.exp(h*c/(i*k_B*T))-1)))
    #plot wavelengths vs fluxes
    
    plt.plot(wavelengths, fluxes)
    plt.title(title + " : Black Body Curve (T= "+ str(int(T))+ "K, R= " + str(R/(7.0*10**10)) + r' $R_\odot$' +")" )
    plt.xlabel(r'$\lambda$' + " (cm)")
    plt.ylabel(r'$B_\lambda$'+ " (erg/s/cm^2)")
    if logplot==True:
        plt.yscale("log")
        plt.xscale("log")
    
```

```python
#function for combined bb spectrum 

def double_bb_lambda(T1, T2, R1, R2,d, l_min, l_max, title, logplot=True):
    #define a range of wavelengths to plot between l_min and l_max
   
    #set step size assuming were plotting in microns
    wavelengths = np.arange(l_min, l_max, 0.01*10**(-4))
    
    #init an empty list for fluxes:
    fluxes = []
    #for loop to calculate a flux for each wavelength:
    for i in wavelengths:
        #define a scaling factor for distance from source:
        dist_scale1 = (R1/d)**2 
        dist_scale2 = (R2/d)**2
        
        #define a flux per v and T, for an object with radius R, at a distance d
        fluxes.append(dist_scale1*(2*h*c**2/i**5)/(np.exp(h*c/(i*k_B*T1))-1)+dist_scale2*(2*h*c**2/i**5)/(np.exp(h*c/(i*k_B*T2))-1))
    #plot wavelengths vs fluxes
    
    plt.plot(wavelengths, fluxes)
    plt.title(title+ " : Black Body Curve" )
    plt.xlabel(r'$\lambda$' + " (cm)")
    plt.ylabel(r'$B_\lambda$'+ " (erg/s/cm^2)")
    if logplot==True:
        plt.yscale("log")
        plt.xscale("log")
    
```

```python
#Question 3d

#import packages to integrate 
import scipy.integrate as integrate

#define constants
R_star = 5 * 7.0*10**10 # in cm
R_c = 1.5*10**(15) #100 au in cm 
#tau_v = v/ (3*10**14)
sigma_b = 5.6*10**(-5) # SB constant in erg/cm^2/s
T_eff = 3000 #K
dist = 3*10**19 # 10 pc in cm


#define bb function 
bb_func = lambda v : ((2 * h * v**3 / c**2) / (np.exp(h * v / (k_B * T_eff)) - 1))*(1-np.exp(-v/(3*10**14)))*(R_star/dist)**2

integral = integrate.quad(bb_func, 10, 10**20)
print(integral)
T_c = (1/(2*sigma_b))*integral[0] 
print(T_c)
```

```python
 #write combined bb function scaled by optical depth
    
def fancy_bb(T1, R1, T2, R2, d, l_min, l_max, title, logplot=True):
    #define a range of wavelengths to plot between l_min and l_max
   #plotting in microns
        wavelengths = np.arange(l_min, l_max, 0.01*10**(-4))

        #init an empty list for fluxes:
        fluxes = []
        #for loop to calculate a flux for each wavelength:
        for i in wavelengths:
            tau_l = 10**(-4)/i

            #define a scaling factor for distance from source:
            dist_scale1 = (R1/d)**2 
            dist_scale2 = (R2/d)**2
            
            flux1 = dist_scale1*(2*h*c**2/i**5)/(np.exp(h*c/(i*k_B*T1))-1)*np.exp(-tau_l)
            flux2 = dist_scale2*(2*h*c**2/i**5)/(np.exp(h*c/(i*k_B*T2))-1)*(1-np.exp(-tau_l))

            #define a flux per v and T, for an object with radius R, at a distance d
            fluxes.append(flux1+flux2)
            
        #plot wavelengths vs fluxes
        plt.plot(wavelengths, fluxes)
        plt.title(title+ " : Black Body Curve scaled by "+ r'$\tau_\nu$')
        plt.xlabel(r'$\lambda$' + " (cm)")
        plt.ylabel(r'$B_\lambda$'+ " (erg/s/cm^2)")
        if logplot==True:
            plt.yscale("log")
            plt.xscale("log")
```

```python
#Plot scaled and unscaled binaries as well as individual components
bb_lambda(T_eff,  R_star,dist ,10**(-6), 1,  "Proto-Star")
bb_lambda(30.0, R_c, dist,10**(-6), 1,  "Cloud")
fancy_bb(T_eff, R_star, 30.0, R_c, dist, 10**(-6), 1,  "Proto-Star in Cloud")
double_bb_lambda(T_eff, 30.0, R_star, R_c ,dist, 10**(-6), 1,  "Proto-Star in Cloud binary")
plt.legend(['Proto-Star', 'Cloud', r'Proto-Star in Cloud w/ $\tau_\nu$', r'binary w/o $\tau_\nu$'])
#set limits on y axis to get a better look at the long wavelength features
plt.ylim(10**(-61), 10**(18))

```
