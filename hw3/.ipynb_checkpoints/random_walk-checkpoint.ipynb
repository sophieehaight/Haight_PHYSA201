```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#define constants
au = 1.5e13
#estimate cross section density to calculate mfp
#calculate number density  with mass/radius of cloud and molecular mass
m_cloud = 2e33 #mass of cloud in g
m_p= 1e-24 #mass of a proton for HII in g
N= m_cloud/m_p
R_cloud = 3.6e15 #0.001pc converted to cm 
V_cloud = 4/3 * np.pi* R_cloud**3
n = N/V_cloud #number density in # cm^-3
sigma = 6.65e-25 # Thompson cross section in cm^2
mfp = 1/(n*sigma)

print("mfp =" + str(mfp/au) +" AU")
```

```python

# asked Chat GPT to generate a random walk inside a spherical shell and then moved arounbd parts of the code so it was structured in a way that made sense to me

#define random walk function 
def random_walk(num_steps, shell_radius):
    
    positions = [np.zeros(3)]  # Start at the center of the sphere

    for _ in range(num_steps):

        #define random step
        direction = np.random.normal(size=3)  #drawing three numbers from a random distribution to define a vector
        norm = np.linalg.norm(direction)      #normalizing the array to some standard step size
        
        step = direction *mfp / norm               #define step as normalized vector multiplied by the mpf of HII
        #print(step/au)

            
        #update position
        new_position = positions[-1] + step
        #print(new_position/au)
        
        # Check if the walker has reached the edge of the cloud
        distance_to_origin = np.linalg.norm(new_position)
        #print(distance_to_origin/au)

        if distance_to_origin >= shell_radius:
            break
        
        # else move the walker to the new position
        positions.append(new_position)


    return np.array(positions)
        


# output a list of positions 
positions = random_walk(1000, R_cloud)

```

```python
# Here I asked Chat GPT to make a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x_coords = positions[:, 0]
y_coords = positions[:, 1]
z_coords = positions[:, 2]


# Plot the path
ax.plot(x_coords/au, y_coords/au, z_coords/au, lw=2)

# Plot the starting point
ax.scatter(x_coords[0]/au, y_coords[0]/au, z_coords[0]/au, color='green', label='Start')
# Plot the ending point
ax.scatter(x_coords[-1]/au, y_coords[-1]/au, z_coords[-1]/au, color='red', label='End')

ax.set_xlabel('X position (au)')
ax.set_ylabel('Y position (au)')
ax.set_zlabel('Z position a')
ax.legend()

plt.show()
```

```python
#interpolate over 1000 simulated paths to accummulate an intensity density
interp = 1000 #number of interpolations 

#repeat random walk simulation and combine list of positions it outputs
all_positions = []
for i in range(interp):
    all_positions.extend(random_walk(1000, R_cloud))  

#this is annoying but I had to norm all the positions and scale by AU
normed_positions= []
for position in all_positions:
    normed_positions.append(np.linalg.norm(position)/au)
    
```

```python
# Calculate the volume of the spherical shell for each bin
volumes = []
for p in normed_positions:
    volumes.append((4/3) * np.pi * ((p/au)**3)) #again, radii are scaled by AU

# Plot the histogram with weighted counts
plt.hist(normed_positions, bins=100, weights=np.ones(len(volumes)) / volumes, edgecolor='black')
plt.xlabel('Radius (AU)')
plt.ylabel('Photon Count/Volume')
plt.yscale('log')
plt.title('Intensity Density -- Photons per Radius Scaled by Volume ')
plt.grid(True)
plt.show()
```

```python
#Run a number of simulations and combine the final position of the scattered particle 
interp=1000 #number of simulations 
last_scatt=[]
for _ in range(interp):
    scatt=random_walk(1000, R_cloud)[-1] #just append final position
    last_scatt.append(np.linalg.norm(scatt) /au) 
   
```

```python
#Plot the distribution of light ray paths 

plt.hist(last_scatt, bins= 100)
plt.title("Total Path Distance Distribution")
plt.xlabel("Position (in units of AU)")
plt.ylabel('Count')
```

```python
#Plot the Density oif the last scatter

# Calculate the volume of the spherical shell for each bin
ls_volumes = []
for p in last_scatt:
    ls_volumes.append((4/3) * np.pi * ((p/au)**3)) #again, radii are scaled by AU

#plot a radial histogram of last scatterings
plt.hist(last_scatt, bins=100, weights=np.ones(len(ls_volumes)) / ls_volumes, edgecolor='black')
plt.xlabel('Radius (AU)')
plt.ylabel('Photon Count/Volume')
plt.xlim(220, 245)
plt.ylim(0, 10e33)
plt.title('Density of Last Scatter -- Photons per Radius Scaled by Volume ')
plt.grid(True)
plt.show()
```
